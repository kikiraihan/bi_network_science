```python
from pathlib import Path
from fastapi.templating import Jinja2Templates
import os, sys
from jinja2 import Environment, FileSystemLoader

sys.path.append("../")
from helper.PlotlyFigHelpers import fig_bar, fig_line, fig_pie,card_template_html_string
```

```python
# Tentukan lokasi direktori template
file_loader = FileSystemLoader(Path("../app/templates"))
env = Environment(loader=file_loader)

# Muat template
template = env.get_template('dashboard_template.html')

# muat card
n_node = 6357 
n_edge = 5177
n_partition = 1843
card_1 = card_template_html_string(h4_title='Jumlah Node', p_subtitle='', id_nya='card_1', icon='bx bxs-network-chart', 
placeholder=n_node, sup_subtitle='Node (Akun)')
card_2 = card_template_html_string(h4_title='Jumlah Edge', p_subtitle='', id_nya='card_2', icon='bx bxs-network-chart', 
placeholder=n_edge, sup_subtitle='Edge (Transaksi)')
card_3 = card_template_html_string(h4_title='Jumlah Komunitas', p_subtitle='', id_nya='card_3', icon='bx bx-group', placeholder=n_partition, sup_subtitle='Partisi Komunitas')

# muat graf
with open('graph_plotly.html', 'r') as file:
    graph_plotly = file.read()

# # muat graf
# with open('graph_plotly_com.html', 'r') as file:
#     graph_plotly_com = file.read()

# muat line com
with open('line_partition.html', 'r') as file:
    line_partition = file.read()

# muat bar from
with open('bar_degree_from.html', 'r') as file:
    bar_degree_from = file.read()

# Muat bar to
with open('bar_degree_to.html', 'r') as file:
    bar_degree_to = file.read()

# Muat bar payment
with open('bar_payment_currency.html', 'r') as file:
    bar_payment_currency = file.read()

# Muat bar receiving
with open('bar_receiving_currency.html', 'r') as file:
    bar_receiving_currency = file.read()

# muat bar_payment_format
with open('bar_payment_format.html', 'r') as file:
    bar_payment_format = file.read()

# muat line date
with open('line_transaction_date.html', 'r') as file:
    line_transaction_date = file.read()



# Render template dengan data
output = template.render(
    name='World', 
    card_1=card_1,
    card_2=card_2,
    card_3=card_3,
    graph_plotly=graph_plotly,
    # graph_plotly_com=graph_plotly_com,
    line_partition=line_partition,
    bar_degree_from=bar_degree_from,
    bar_degree_to=bar_degree_to,
    bar_payment_currency=bar_payment_currency,
    bar_receiving_currency=bar_receiving_currency,
    bar_payment_format=bar_payment_format,
    line_transaction_date=line_transaction_date,
)

file_path_output="../app/pages/index.html"
# Simpan hasil render ke file HTML
with open(file_path_output, 'w') as file:
    file.write(output)
```
